```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
data = load_iris()
x = pd.DataFrame(data.data, columns=[data.feature_names])
y = pd.Series(data.target)
```

```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
```

```python
min_samples_split = np.array([2, 3, 4, 5, 6])
max_depth = np.array([1, 2, 3, 4, 5, 6])
criterion = ['gini', 'entropy']
grid_tests = {
    'min_samples_split': min_samples_split,
    'max_depth': max_depth,
    'criterion': criterion
}
```

```python
model = DecisionTreeClassifier()
grid = GridSearchCV(estimator=model, param_grid=grid_tests, cv=5)
grid.fit(x, y)

print(f'The best score of this model was: {grid.best_score_}. The best parameters of this model was: {grid.best_params_}')
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
```

```python
best_model = DecisionTreeClassifier(criterion='gini', max_depth=3, min_samples_split=2)
best_model.fit(x, y)

plt.figure(figsize=(10, 8))
plot_tree(best_model, feature_names=data.feature_names)
plt.show()
```
