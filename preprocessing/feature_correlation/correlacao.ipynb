```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = '/Users/leona/OneDrive/Desktop/MODULO1_MACHINE_LEARNING/pre-processamento/2015-building-energy-benchmarking.csv'
```

```python
df = pd.read_csv(data)
```

```python
for i in df.columns:
    if df[i].dtype == object:
        df.drop(columns=i, axis=1, inplace=True)
df.dtypes
```

```python
for i in df.columns:
    df[i].fillna(df[i].mean(), inplace=True)
```

```python
display(df.corr(method='pearson'))
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(df.corr())
plt.show()
```
